# Ejemplos de cartas

Cuando podemos escribir el conjunto abierto de la variedad 2D como $x^3(x^1, x^2)$, una parametrización "trivial" viene dada por la proyección $\pi(x^1, x^2, x^3) = (x^1, x^2)$. En los ejemplos interactivos de abajo, los grids sobre las superficies representan las curvas $x^1=\text{constante}$ y $x^2=\text{constante}$.

```python
# Superficies con coordenadas dadas por la proyección "trivial"

x1 = np.reshape(np.linspace(-5, 5, 101), (1, -1))
x2 = np.reshape(np.linspace(-5, 5, 101), (-1, 1))

x3 = dict()
x3["paraboloide hiperbolico"] = x1**2 - x2**2
x3["paraboloide eliptico"] = x1**2 + x2**2
x3["oscilaciones radiales"] = np.cos(np.sqrt(x1**2 + x2**2))**2 / np.sqrt(x1**2 + x2**2 + 5e-1)

fig, axs = plt.subplots(1, len(x3), subplot_kw={"projection": "3d"}, figsize= (4*len(x3), 4))

for key, ax in zip(x3.keys(), axs):
    ax.plot_surface(x1, x2, x3[key], edgecolor='royalblue', lw=0.5, rstride=10, cstride=10, alpha=0.5)
    ax.set_title(key)
```

En los ejemplos siguientes, representamos las superficies (esfera y toro) a través de su parametrización en términos de dos ángulos, $\theta$ y $\phi$. En consecuencia, las curvas dibujadas sobre las superficies se corresponden con $\theta=:x^1=\text{constante}$ y $\phi=:x^2=\text{constante}$.

```python
# Esfera y toro

θ = np.reshape(np.linspace(0, np.pi, 31), (1, -1))
ϕ = np.reshape(np.linspace(0, 2 * np.pi, 61), (-1, 1))

x1 = dict()
x2 = dict()
x3 = dict()

x1["esfera"] = np.sin(θ) * np.cos(ϕ) 
x2["esfera"] = np.sin(θ) * np.sin(ϕ) 
x3["esfera"] = np.cos(θ) * np.ones(ϕ.shape)

x1["toro"] = np.cos(2 * θ) * (0.8 + 0.4 * np.cos(ϕ))
x2["toro"] = np.sin(2 * θ) * (0.8 + 0.4 * np.cos(ϕ))
x3["toro"] = 0.4 * np.sin(ϕ)

fig, axs = plt.subplots(1, len(x3), subplot_kw={"projection": "3d"}, figsize= (12, 4))

for key, ax in zip(x3.keys(), axs):
    ax.plot_surface(x1[key], x2[key], x3[key], edgecolor='royalblue', lw=0.5, rstride=3, cstride=3, alpha=0.3)
    ax.set_title(key)
    ax.set_xlim(-1.4, 1.4)
    ax.set_ylim(-1.4, 1.4)
    ax.set_zlim(-1.4, 1.4)
```

```python
# esfera: parametrizaciones alternativas

x1p = np.reshape(np.linspace(-5, 5, 101), (1, -1))
x2p = np.reshape(np.linspace(-5, 5, 101), (-1, 1))

x1 = dict()
x2 = dict()
x3 = dict()

x1["esfera proyectiva"] = 2 * x1p / (1 + x1p**2 + x2p**2)
x2["esfera proyectiva"] = 2 * x2p / (1 + x1p**2 + x2p**2)
x3["esfera proyectiva"] = (x1p**2 + x2p**2 - 1) / (1 + x1p**2 + x2p**2)

x1p = np.reshape(np.linspace(-1, 1, 101), (1, -1))
x2p = np.reshape(np.linspace(-1, 1, 101), (-1, 1))

x1["hermisferio norte cartesiano"] = x1p 
x2["hermisferio norte cartesiano"] = x2p
x3["hermisferio norte cartesiano"] = np.sqrt(1 - x1p**2 - x2p**2)

fig, axs = plt.subplots(1, len(x3), subplot_kw={"projection": "3d"}, figsize= (12, 4))
strides = [4, 10]

for s, key, ax in zip(strides, x3.keys(), axs):
    ax.plot_surface(x1[key], x2[key], x3[key], edgecolor='royalblue', lw=0.5, rstride=s, cstride=s, alpha=0.3)
    ax.set_title(key)
    ax.set_xlim(-1.4, 1.4)
    ax.set_ylim(-1.4, 1.4)
    ax.set_zlim(-1.4, 1.4)
```
